```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('OnlineRetail.csv',encoding='unicode_escape') #unicode_escape for unknown language escape
df
```

```python
df.dtypes
```

```python
df.sort_values(by='InvoiceDate', ascending=False)
```

```python
current_date = pd.to_datetime('01-01-2012')
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df
```

```python
df.dtypes
```

```python
max_date = df.groupby(['CustomerID']).max()[['InvoiceDate']]
max_date
```

```python
max_date['InvoiceAge'] = (current_date - max_date['InvoiceDate']).dt.days
max_date
```

```python
#Last shopping date of a customer
recency=max_date.drop('InvoiceDate',axis=1)
recency
```

```python
frequency=df.drop_duplicates('InvoiceNo')
frequency
```

```python
#frequency
frequency= frequency.groupby(['CustomerID']).count()[['InvoiceNo']]
#frequency = frequency.groupby('CustomerID')['InvoiceNo'].count().reset_index()
#frequency.rename(columns={'InvoiceNo': 'InvoiceCount'}, inplace=True)
#frequency = frequency.reset_index(drop=True)
#frequency = frequency.sort_values(by='InvoiceCount', ascending=False)
frequency
```

```python
#monetary
df['Total']= df['Quantity']*df['UnitPrice']
df
```

```python
df[df.CustomerID ==12346.0]
```

```python
money = df.groupby(['CustomerID'])[['Total']].sum()
money
```

```python
RFM=pd.concat([recency,frequency,money], axis=1)
RFM
```

```python
RFM.columns=['Recency', 'Frequency', 'Monetary']
RFM
```

```python
RFM.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled =scaler.fit_transform(RFM)
```

```python
scaled
```

```python
scaled.shape
```

```python
from sklearn.cluster import KMeans
```

### To find out the value of K

```python
k_range = range(1,10)
sse=[]
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(scaled)
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_range,sse,'-r*')
plt.xlabel('K')
plt.ylabel('Sum of square error')
plt.title('Elbo theory')
```

### Object Creation and data fitting

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(scaled)
```

```python
kmeans.labels_
```

```python
kmeans.labels_.shape
```

```python
RFM['Clusters'] = (kmeans.labels_+1)
```

```python
#RFM['Clusters'].unique()
RFM
```

```python
RFM[RFM.Clusters==2]
```

```python
RFM[RFM.Clusters==1]
```

```python
RFM[RFM.Clusters==3]
```

```python
final_result = RFM.groupby(['Clusters']).mean()[["Recency","Frequency","Monetary"]]
final_result
```

```python
def rank(row):
    if row['Clusters']==3:
        return 'Diamond'
    elif row['Clusters']== 1:
        return 'Gold'
    else:
        return 'Silver'
```

```python
RFM['Group']= RFM.apply(rank,axis=1)
RFM
```

```python
result=RFM.Group.value_counts()
result
```

```python
result.plot(kind="bar", color=["Orange","Red","Green"])
```
